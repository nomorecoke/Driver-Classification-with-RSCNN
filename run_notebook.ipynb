```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
from sklearn.metrics import confusion_matrix
import Generate_Data
from NN import Dense, CNN
NP_SEED = 777
np.random.seed(NP_SEED)
SPLIT_RATIO = 0.75
TRIP = 2
CHANNEL = 5
```

```python
def random_shuffle(x, y):
    p = np.random.permutation(len(x))
    return x[p], y[p]
```

```python
# INPUT : validation_num (trip 0->2,), CHANNEL(4 is basic)
Generate_Data.seperate_train_val(TRIP,CHANNEL)
```

# 전체 데이터 shuffle 하여 사용

```python
# shuffle all data
train_x = np.load("./train_x.npy")
train_y = np.load("./train_y.npy")
val_x = np.load("./val_x.npy")
val_y = np.load("./val_y.npy")

x = np.concatenate((train_x, val_x),axis = 0)
y = np.concatenate((train_y, val_y),axis = 0)


x, y  = random_shuffle(x,y)
train_x = x[:int(len(x) *SPLIT_RATIO)]
val_x = x[int(len(x) *SPLIT_RATIO):]
train_y = y[:int(len(x) *SPLIT_RATIO)]
val_y = y[int(len(x) *SPLIT_RATIO):]

test_dict = np.load("./test_dict.npy")
test_dict = test_dict.item()

data_input = {"train":{"x":train_x, "y":train_y},
              "val":{"x":val_x, "y":val_y},
             "test":test_dict}


print(train_x.shape, train_y.shape,val_x.shape,val_y.shape)

```

# TRIP 단위로 FOLD 할때 사용하는 CELL

```python
train_x = np.load("./train_x.npy")
train_y = np.load("./train_y.npy")
val_x = np.load("./val_x.npy")
val_y = np.load("./val_y.npy")
train_x, train_y = random_shuffle(train_x, train_y)
val_x, val_y = random_shuffle(val_x, val_y)

test_dict = np.load("./test_dict.npy")
test_dict = test_dict.item()

data_input = {"train":{"x":train_x, "y":train_y},
              "val":{"x":val_x, "y":val_y},
             "test":test_dict}

print(train_x.shape, train_y.shape,val_x.shape,val_y.shape)
```

# TRAIN

```python
tf.reset_default_graph()
sess = tf.Session()
BATCH_SIZE = 64
EPOCH = 200
LEARNING_RATE = 1e-4

nn = CNN(sess,46,CHANNEL,0.5,LEARNING_RATE, 0.9,"Dense")
sess.run(tf.global_variables_initializer())
nn.train(data_input, BATCH_SIZE, EPOCH)
```
